## Convert cubic and Cylinder and split into interpolatin and extrapolation

```python
import pandas as pd

# Define the function to convert cube strength to cylinder strength
def convert_strength_to_cylinder(strength, specimen_shape):
    """
    Convert cube strength to cylinder strength using a factor of 0.8 for normal weight concrete.
    """
    if specimen_shape.lower() == 'cube':
        return strength * 0.8
    return strength

# Load the data
file_path = "transformed_data Descriptive.xlsx"  # Adjust file path if necessary
df = pd.read_excel(file_path)

# List of columns to keep
columns_to_keep = [
    "SiO2 %",
    "Al2O3 %",
    "Fe2O3 %",
    "CaO %",
    "OPC - content in mix -  (kg/m3)",
    "FA - content in mix -  (kg/m3)",
    "GGBFS - content in mix -  (kg/m3)",
    "SF - content in mix -  (kg/m3)",
    "MK - content in mix -  (kg/m3)",
    "Other SCM - content in mix -  (kg/m3)",
    "Total aggregates - content - (kg/m3)",
    "Water - content from all sources in mix - (kg/m3)",
    "SiO2 - liquid percentage in Na2SiO3 solution - %",
    "SiO2 -Dry content from Na2SiO3 solution in mix - (kg/m3)",
    "NaOH solution - content in mix - (kg/m3)",
    "NaOH solution - concentration - molar",
    "Superplasticizer - content - (kg/m3)",
    "Extra water - content in mix - (kg/m3)",
    "Initial Curing Process - duration - (day)",
    "Initial Curing Process - Temperature - (ºC)",
    "Test age",
    "Specimen shape",
    "Cylinder Height - length - (mm)",
    "Cylinder Diameter - length - (mm)",
    "Cube Edge - length - (mm)",
    "Na2O -Dry content from Na2SiO3 solution in mix - (kg/m3)",
    "Binder - Fine modulus - (m2/kg)",
    "Compressive test"
]

# Select the specified columns from the dataframe
df_selected = df[columns_to_keep]

# Convert compressive strength values based on specimen shape
df_selected['Compressive test'] = df_selected.apply(
    lambda row: convert_strength_to_cylinder(row['Compressive test'], row['Specimen shape']),
    axis=1
)

# Calculate the 80% quantile of the updated 'Compressive test' column
quantile_80 = df_selected["Compressive test"].quantile(0.8)

# Split the data below the 80% quantile into 75% training and 25% testing (interpolation)
df_below_quantile = df_selected[df_selected["Compressive test"] <= quantile_80]
df_interp_train = df_below_quantile.sample(frac=0.75, random_state=1)
df_interp_test = df_below_quantile.drop(df_interp_train.index)

# The data above the 80% quantile is used for extrapolation testing
df_extrap_test = df_selected[df_selected["Compressive test"] > quantile_80]

# Output the sizes of the resulting datasets
print("Interpolation Training Set Size:", len(df_interp_train))
print("Interpolation Test Set Size:", len(df_interp_test))
print("Extrapolation Test Set Size:", len(df_extrap_test))

# Optional: Save the resulting dataframes if needed
df_interp_train.to_excel("cubic_to_cylinder_intra_extrapolation/interpolation_train.xlsx", index=False)
df_interp_test.to_excel("cubic_to_cylinder_intra_extrapolation/interpolation_test.xlsx", index=False)
df_extrap_test.to_excel("cubic_to_cylinder_intra_extrapolation/extrapolation_test.xlsx", index=False)
```

## Filter dataset based on Binder

### FLy Ash and Slag

```python
# Load the Excel file for refining the "Fly Ash + Slag" dataset
file_path_fly_ash_and_slag = 'cubic_to_cylinder_intra_extrapolation/extrapolation_test.xlsx'
excel_data_fly_ash_and_slag = pd.ExcelFile(file_path_fly_ash_and_slag)

# Parse the relevant sheet
data_fly_ash_and_slag = excel_data_fly_ash_and_slag.parse('Sheet1')

# Define conditions to strictly exclude additional binders for "Fly Ash + Slag"
fly_ash_and_slag_condition = (
    data_fly_ash_and_slag['MK - content in mix -  (kg/m3)'].isna() &  # Ensure no MK
    data_fly_ash_and_slag['OPC - content in mix -  (kg/m3)'].isna() &  # Ensure no OPC
    data_fly_ash_and_slag['SF - content in mix -  (kg/m3)'].isna() &  # Ensure no Silica Fume
    data_fly_ash_and_slag['Other SCM - content in mix -  (kg/m3)'].isna()  # Ensure no other SCMs
)

# Apply the filter to get only those rows that have both Fly Ash and Slag without any other binders
filtered_fly_ash_and_slag = data_fly_ash_and_slag[
    (data_fly_ash_and_slag['FA - content in mix -  (kg/m3)'].notna()) &
    (data_fly_ash_and_slag['GGBFS - content in mix -  (kg/m3)'].notna()) &
    fly_ash_and_slag_condition
]

# Save the refined dataset for "Fly Ash + Slag"
output_file_path_fly_ash_and_slag = 'cubic_to_cylinder_intra_extrapolation/filter_based_on_binder/extrapolation_test/fly_ash_and_slug_extrapolation_test.xlsx'
filtered_fly_ash_and_slag.to_excel(output_file_path_fly_ash_and_slag, index=False)

# Display the refined file path for user reference
output_file_path_fly_ash_and_slag
```

## Only Slag

```python
# Load the Excel file for refining the "Only Slag" dataset
file_path_slag = 'cubic_to_cylinder_intra_extrapolation/extrapolation_test.xlsx'
excel_data_slag = pd.ExcelFile(file_path_slag)

# Parse the relevant sheet
data_slag = excel_data_slag.parse('Sheet1')

# Define conditions to strictly exclude additional binders for "Only Slag"
only_slag_condition = (
    data_slag['FA - content in mix -  (kg/m3)'].isna() &  # Ensure no Fly Ash
    data_slag['MK - content in mix -  (kg/m3)'].isna() &  # Ensure no MK
    data_slag['OPC - content in mix -  (kg/m3)'].isna() &  # Ensure no OPC
    data_slag['SF - content in mix -  (kg/m3)'].isna() &  # Ensure no Silica Fume
    data_slag['Other SCM - content in mix -  (kg/m3)'].isna()  # Ensure no other SCMs
)

# Apply the filter to get only those rows that have Slug (GGBFS) without any other binders
filtered_slag_only = data_slag[(data_slag['GGBFS - content in mix -  (kg/m3)'].notna()) & only_slag_condition]

# Save the refined dataset for "Only Slag"
output_file_path_slag = 'cubic_to_cylinder_intra_extrapolation/filter_based_on_binder/extrapolation_test/only_slag_extrapolation_test.xlsx'
filtered_slag_only.to_excel(output_file_path_slag, index=False)

# Display the refined file path for user reference
output_file_path_slag
```
